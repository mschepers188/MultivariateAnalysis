```python
# import modules
import pandas as pd
from pydataset import data
from sklearn.decomposition import FactorAnalysis
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data prep
df = data('bioChemists')
df=df.iloc[1:250]
# We don't want categorical variables for the analysis
X=df[['art', 'kid5', 'phd', 'ment']]
X
```

```python
# Specify the number of desired factors
fact_2c=FactorAnalysis(n_components=2)
# Takes dataset X and created the factors by performing FA
X_factor=fact_2c.fit_transform(X)
X_factor
```

```python
thisdict = {'Single':'0', 'Married':'1'}
# Set categories to integers for plotting
z = np.array(df.mar.map(thisdict), dtype=np.int8)
# make color array for plotting
colors = np.array(['yellow', 'purple'])
```

```python
plt.scatter(X_factor[:, 0], X_factor[:, 1], c=colors[z])
plt.show()
# Terrible separation, most likely not a relevant factor in marriage
```
