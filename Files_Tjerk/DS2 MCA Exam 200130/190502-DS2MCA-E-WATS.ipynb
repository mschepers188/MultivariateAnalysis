### Assignment 3: MultiDimensional Scaling (Principal Coordinate Analysis).

Execute the code below (no points for that...)

```python
S = data.iloc[::10, 1:].values
D = ((S[:, None, :] - S[None, :, :])**2).sum(axis=2)
m = D.mean(axis=1)
D = -0.5 * (D - m[:, None] - m[None, :] + m.mean())
pv, pcoor = np.linalg.eig(D)

plt.plot(range(1, 11), pv[:10])
plt.show()

plt.scatter(pcoor[:,0], pcoor[:,1], c=pcoor[:,2], s=1)
plt.show()
```
